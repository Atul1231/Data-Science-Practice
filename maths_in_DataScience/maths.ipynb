```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import skewnorm
```

```python
a= 10
size =10000
data = np.random.normal(loc=0,scale=1,size=10000)
data2 = skewnorm.rvs(a,size=size)
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

#plot data on each subplot
axs[0].hist(data)
axs[0].set_title('Normal Distribution')
axs[1].hist(data2)
axs[1].set_title('Skewnorm Distribution')

plt.show()
```

```python
sorted_data = np.sort(data)
sorted_data2 = np.sort(data2)

#calculate the percentiles
n=len(sorted_data)
percentiles=(np.arange(1,n+1) - 0.5) / n * 100

#plot the quantile plot
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].plot(sorted_data,percentiles, marker='o',linestyle='none')
ax[1].plot(sorted_data2,percentiles,marker='o',linestyle='none')
plt.grid(True)
plt.show()
```

```python
data = np.random.normal(loc=0,scale=1,size=1000)
fig,ax=plt.subplots()
stats.probplot(data,dist='norm' , plot=ax)
ax.get_lines()[1].set_color('red')
ax.get_lines()[1].set_linestyle('dashed')
ax.get_lines()[1].set_linewidth(2)

plt.title('QQ Plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid(True)
plt.show()
```
