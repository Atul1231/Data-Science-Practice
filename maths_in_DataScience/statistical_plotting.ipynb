```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
#uniform distribution
plt.hist(stats.uniform.rvs(size=10000),edgecolor='black',color='lightgreen')
plt.title('Uniform Distribution')
plt.ylabel('Number of Observations')
plt.show()
```

```python
#poisson distribution
fig,axs=plt.subplots(4)
fig.suptitle('Poisson Distribution')
fig.tight_layout(rect=[0,0,1.05,0.95])
color_list=['r','g','b','purple']
bins = np.arange(11) - .5

for i,j in zip(range(4),color_list):
    axs[i].hist(stats.poisson.rvs(mu=i+1,size=100),bins=bins,edgecolor='black',color=j,label=f'$\lambda={i+1}$')
    
fig.legend(bbox_to_anchor=(1, 0.92))
```

```python
#normal distribution
mean = 0
variance = 1
std_dev = variance ** 0.5
norms_sample=np.random.normal(size=10000)

plt.hist(norms_sample, density=True , edgecolor='black', color='lightblue',label='$\overline{x}=$' + f'{np.mean(norms_sample):.3f}$' + f'{np.std(norms_sample):.3f}$')
xmin, xmax = plt.xlim()
x= np.linspace(mean - 4*std_dev, mean + 4*std_dev, 10000)
p= stats.norm.pdf(x, mean, std_dev)
plt.plot(x, p, linewidth=2,color='purple',label='standard normal distribution')
plt.legend()
```
