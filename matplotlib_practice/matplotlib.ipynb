```python
#key terms before learning matplotlib
#data point-----n(x,y)
#x-axis(input) and y-axis(output)
#figure (whole window)
#axes(limited window and only the box of the graph)
#plot(actual graph)
#marker(symbol to make the data points visible)
#line style (styling of a line --:)
#color
#legend(description of the graph and which line denots what in graph)
#label(used when we need to explain what axes are doing)
#title (title of a graph)
#grid(background lines in a graph)
#function(block of code do specific task)
#method(when in object we need to declare some task)
#parameters(given like what they will do for controlling)
#keyword arguments (value for the para) like color = green
#oo api
#dpi higher dpi more clear images
#backend
#use object oriented api instead of pyplot(state variable) when working with large data (pyplot is global set) aand oo a is controlled by us
#ooa is more flexible than pyplot in bigger datasets
#ooa is perfect for begginer but heavily recommend to use ooa when in industry
```

```python
#pyplot
import matplotlib.pyplot as plt
x=['mon','tue','wed','thurs','fri']
y=[12,3,49,30,20]

plt.plot(x,y)
plt.title('Bakery sales per days')

plt.xlabel('Days of week')
plt.ylabel('Sales on the day')
plt.show()
```

```python
#functions in pyplot
#plt.plot(x, y, color='color name' , linestyle='line_Style' , linewidth='value' , marker='' , label='')
import matplotlib.pyplot as plt

months=[1,2,3,4,5]
sales=[10000,30000,58882,37433,90000]

plt.plot(months,sales,color='red',linestyle='--',marker='o',linewidth='2',label='sales in 2025')
plt.xlabel('Months')
plt.ylabel('sales per month')
plt.title('Monthly sales data report')
plt.legend(loc='lower right',fontsize=12)
plt.grid(color='gray',linestyle=':',linewidth=1)
plt.xlim(1,5)
plt.ylim(1000,100000)
plt.xticks([1,2,3,4,5],['A1','A2','A3','A4','A5'])
plt.show()
```

```python
#plt.bar(x,height,color='color name ',width=value,label='label name')
import matplotlib.pyplot as plt
product=['A','B','C','D',]
sales=[1000,10000,2000,8000]

plt.bar(product,sales,color='pink',label='sales per product')
plt.xlabel('products')
plt.ylabel('sales')
plt.title('Sales per product')
plt.legend()
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
regions=['North','South','East','West']
literacy=[10,1,9,9]

plt.pie(literacy,labels=regions,autopct='%1.1f%%',colors=['red','yellow','green','orange'])
plt.title('100% literacy rate sir')
plt.show()
```

```python
#histogram
import matplotlib.pyplot as plt
scores=[45,37,33,42,53,38,83,95,60,63,79,63,79]
plt.hist(scores,bins=5,color='gold',edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of students')
plt.title('Score Distributions of students')
plt.show()
```

```python
#Scatter graph : uses in defining the relation between 2 variables
import matplotlib.pyplot as plt
hours_studies=['2','3','10','1','2']
exam_scores=['20','80','30','70','50']

plt.scatter(hours_studies,exam_scores,color='pink',marker='o',label='Student Data')
#plt.scatter([1,2,3,4,5],[30,59,50,70,80],color='pink',marker='o',label='Class A')
#plt.scatter([1,2,3,4,5],[30,59,50,70,80],color='pink',marker='o',label='Class B')
#in the above input it will give the comparison of two classes a and b
plt.xlabel('hours studied')
plt.ylabel('exam scores')
plt.title('students scores to hours data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#subplot functions
#multiple charts
#space efficent
#implement quickly


#using subplot(nrows,ncols,index)

import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,20,30,40,50]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('line chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('bar chart')

plt.show()
```

```python

import matplotlib.pyplot as plt

#fig, ax=subplots(nrows,ncols,figsize=(width,height))
fig,ax=plt.subplots(1,2,figsize=(10,4))
x=[1,2,3,4,5]
y=[10,20,30,40,50]

ax[0].plot(x,y)
ax[0].set_title('Line plot')

ax[1].bar(x,y)
ax[1].set_title('Bar plot')

plt.tight_layout()
plt.show()
```
